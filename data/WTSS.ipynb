```python
from wtss import wtss
import geopandas as gpd
import pandas as pd
```

```python
w = wtss("http://www.esensing.dpi.inpe.br")

cv_list = w.list_coverages()
```

```python
cv_list['coverages'][0]
```

```python
shape = gpd.read_file('./shapes/para_2018/regiao_para_2018.shp')
shape.plot()
```

```python
lat = shape['Latitude'].values
lon = shape['Longitud'].values
```

```python
a = w.time_series(coverage=cv_list['coverages'][0], attributes='Precipit', latitude=lat[0], longitude=lon[0])
```

```python
a.attributes
```

```python
a.attributes['ndvi']
a.timeline
df = pd.DataFrame({'ndvi': a.attributes['ndvi'], 'datetime': a.timeline})
df.head()
```
